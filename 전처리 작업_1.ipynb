```python
# 등급 범주화
def rating(x):
    if x=="TV-MA" or x=="R" or x=="NC-17": 
        return 0 # 18세 미만 관람불가
    elif x=='TV-14' or x=='PG-13':
        return 1 # 14세 미만 관람불가
    elif x=='TV-PG' or x=='TV-G' or x=="PG":
        return 2 # 어린이 시청시 보호자 지도
    elif x=='TV-Y7-FV' or x=='TV-Y7':
        return 3 # 7세 미만 관람불가
    elif x=="TV-Y":
        return 4 # 영유아 프로그램
    elif x=="NR" or x=="UR":
        return 5 # 등급 받지 않음
    else:
        return 6 # 전체 관람
    
data["rating"]=data["rating"].apply(rating)
```

```python
data["rating"].unique()
data["rating"].value_counts()
```

```python
# 평점 범주화 - 4이하면 0, 5이상이면 1. 즉, 0은 낮은 평점을 받음. 1은 높은 평점을 받음
def test(x):
    if x<=4: 
        return 0
    else: 
        return 1
        
data["U44_F_star"] = data["U44_F_star"].apply(test)
    
```
